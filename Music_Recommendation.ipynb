```python
categorical = ['source_system_tab','source_screen_name','source_type']
for i in categorical:
    mode=data_train[i].mode()[0]
    data_train[i].fillna(mode,inplace=True)
```

```python
source_screen_name_dummies = pd.get_dummies(data_train['source_screen_name'])
```

```python
source_type_dummies = pd.get_dummies(data_train['source_type'])
```

```python
source_type_dummies
```

```python
X_train = pd.concat([source_system_tab_dummies,source_screen_name_dummies,source_type_dummies],axis=1)
```

```python
X_train
```

```python
y_train = data_train['target']
```

```python
y_train
```

## **We will apply BernuulliNB to classiy our binary target.::**

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
```

```python
nb_classifier = BernoulliNB()

# Train the model
nb_classifier.fit(X_train, y_train)
```

```python
y_pred=nb_classifier.predict(X_train)
```

```python
#train accuracy
accuracy = accuracy_score(y_train, y_pred)
print("Accuracy:", accuracy)
```
